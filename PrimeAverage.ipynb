```python
%%writefile prime_average.py

from mrjob.job import MRJob
import numpy as np
import os

class PrimeAverage(MRJob):
    row_count = 0
    def mapper(self, _, line):
        try:
            numbers = line.split()
            self.row_count += 1
            # yield all the number on the row, the text row in the beginning is ignored
            for i in map(int, numbers):
                yield self.row_count, i
        except ValueError:
            pass
        
    # because all the samples are of the same size,
    # the mean of means equals the mean of the entire set
    def combiner(self, key, valuelist):
        # Get mean of the row
        length = 0
        sums = 0
        for i in valuelist:
            sums += i
            length += 1 
        yield 0, sums/length
        
    def reducer(self, key, valuelist):
        # Get the mean of all the rows
        listed = list(valuelist)
        yield "Average:", np.mean(listed)
    

if __name__ == "__main__":
    PrimeAverage.run()
```
