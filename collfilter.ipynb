```python
import pandas as pd
data = pd.read_csv("C:\\Users\\Heidi\\Documents\\Introduction to big data\\lastfm-matrix-germany.csv", sep=",", encoding="latin")
data
```

```python
# remove rows with all zeros, ignore id row
data = data[(data[data.columns.values[1:]] != 0).any(axis=1)]
# remove artists with zero listeners (none in the source data)
data = data.loc[:, (data != 0).any(axis=0)]
data
```

```python
from scipy.spatial.distance import pdist, squareform
import numpy as np
no_user = data.drop('user', 1)
art_dists = squareform(pdist(no_user.T, "cosine"))

min_amount = 5
# Get the five most similar artist pairs
def get_n_closest(min_amount, dists):
    min_values = [{"i": -1, "j": -1, "val": 2}]
    for i, row in  enumerate(dists):
        for j, cell in enumerate(row):
            if i == j: # ignore cell when it's the same artist
                continue
            if any(x["i"] == j and x["j"] == i for x in min_values): # if the pair is already in list
                continue
            # if the cell is lower than the lowest values, save it and its indices
            max_val = max(min_values, key=lambda x:x['val'])
            if cell < max_val["val"]:
                min_values.append({"i": i, "j": j, "val": cell})
                # if min values grows bigger than allowed, take out the biggest
                if len(min_values) > min_amount:
                    min_values.remove(max_val)
    return min_values

min_artists = get_n_closest(min_amount, art_dists)

print("The", str(min_amount), "most similar artists")
for x in min_artists:
    art_i = no_user.columns.values[x["i"]]
    art_j = no_user.columns.values[x["j"]]
    print(str(x["val"]) + ": (" + art_i + ", " + art_j + ")")


user_dists = squareform(pdist(no_user, "cosine"))
min_users = get_n_closest(min_amount, user_dists)

print("The", str(min_amount), "most similar users")

for x in min_users:
    user_i = data["user"][x["i"]]
    user_j = data["user"][x["j"]]
    print(str(x["val"]) + ": (" + str(user_i) + ", " + str(user_j) + ")")

artist_list = no_user.columns.values.tolist()

#Most similar artist for Michael Jackson, Madonna and Scooter
def most_similar_for_artist(name, count, dists):
    index = artist_list.index(name)
    # Get k smallest values from artists distances
    idx = np.argpartition(dists[index], count)[:count].tolist()
    idx.remove(index)
    # Print most similar artist
    print("The most similar artist for", name)
    print(no_user.columns.values[idx])

michael = most_similar_for_artist("michael jackson", 10, art_dists)
madonna = most_similar_for_artist("madonna", 10, art_dists)
scooter = most_similar_for_artist("scooter", 10, art_dists)

```
